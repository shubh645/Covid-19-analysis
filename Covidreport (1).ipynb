```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #type ignore
coviddf=pd.read_csv("C:\\Users\\SHUBHAM KADAM\\Downloads\\covid_19_india.csv")
print(coviddf.head())
coviddf.info()
```

```python
df=pd.read_csv("C:\\Users\\SHUBHAM KADAM\\Downloads\\StatewiseTestingDetails.csv")
print(df)
df['Negative']=pd.to_numeric(df['Negative'],errors='coerce')
```

```python
statevisetestingdf=df.dropna()
print(statevisetestingdf.head())
statevisetestingdf.info()
print("Descriptive analysis of Statewise Testing Details:")
statevisetestingdf.describe()
newdf1=statevisetestingdf.transpose()
print(newdf1)
```

```python
plt.Figure(figsize=(10,5))
total_cases=statevisetestingdf.groupby('State')[['TotalSamples','Positive','Negative']].sum().sort_values(by='TotalSamples',ascending=False).head(10)
total_cases.plot(kind='bar',stacked=True,color=['blue','red','green'])
plt.title('Top 10 States Total Covid-19 cases in india')
plt.xlabel("States")
plt.ylabel("Total cases")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
data1=[16081863212,230625655,3687952310,71748086235,89427591,9334413022]
label1=["2020 Total samples","2020 Positive cases","2020 Negative cases","2021 Total samples","2021 Positive cases","2021 Negative cases"]
plt.title('Total peoples suffered by Covid-19 in India')
plt.pie(x=data1,labels=label1,startangle=120)
plt.legend()
plt.show()
```

```python
df3=coviddf.describe()
df6=df3.transpose()
print("Descriptive statistics of the dataset:")
print(df3)
```

```python
plt.figure(figsize=(10,5))
top_statesConfirm=coviddf.groupby('State/UnionTerritory')['Confirmed'].sum().sort_values(ascending=False).head(10)
top_statesConfirm.plot(kind='bar',color='skyblue')
plt.title("Top 10 states of confirmed peoples")
plt.ylabel('Confirmed cases')
plt.xlabel("States")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_statesDeaths=coviddf.groupby('State/UnionTerritory')['Deaths'].sum().sort_values(ascending=False).head(5)
top_statesDeaths.plot(color='orange')
plt.fill_between(top_statesDeaths.index,top_statesDeaths.values,color='orange')
plt.title("Top 5 states in Deaths")
plt.ylabel('Death cases')
plt.xlabel("States")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
top_statesCured=coviddf.groupby('State/UnionTerritory')['Cured'].sum().sort_values(ascending=False).head(15)
top_statesCured.plot(kind='bar',color='pink')
plt.title("Top 15 states in Cured")
plt.ylabel('Cured cases')
plt.xlabel("States")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
